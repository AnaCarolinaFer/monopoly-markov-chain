O objetivo do trabalho é modelar a movimentação de um único jogador no tabuleiro durante uma partda de Banco imobiliário, seguindo regras pré-determinadas, a partir de uma Cadeia de Markov.

Como sabemos, uma variável aleatória é uma variável quantitativa, cujo resultado depende de fatores aleatórios. Já uma cadeia de Markov é uma sequência de variáveis aleatórias X0, X1, X2, ..., Xn onde Xn+1 não depende de X1, X2, ..., Xn-1, caso se conheça Xn, ou seja Xn+1 e Xn não são independetes entre si, isto é, a ocorrência de uma é influenciada pela ocorrência da outra.

Para este caso, como desejamos modelar a movimentação do jogador no tabuleiro, as variaveis aleatórias assumiram valores correspondentes as casas do tabuleiro, já que o resultado de uma movimentação no jogo é o jogador cair em uma certa casa. O tabuleiro modelado consiste em 40 casas, dispostas circularmente, enumeradas de 0 até 39. Com isso, temos 40 elementos no conjunto de estados possiveis desta cadeia.

Além disso, uma vez que uma jogada define uma movimentação do jogador para uma determinada casa do tabuleiro, para que uma nova jogada ocorra é necessario identificar a localização atual do jogador, isto é, o resultado da jogada imediatamente anterior, tornando a próxima jogada dependente da jogada imediatamente anterior, como diz a definição de cadeias de Markov.

Durante uma jogada, o numero de casas que o jogador irá andar, sempre no sentido crescente, é a soma do resultado obtido em dois lançamentos de um dado honesto de 6 faces. Logo, podemos afirmar que existem dois eventos aleatórios que modelam a movimentação do jogador, assim a probabilidade de transição de um estado da cadeia para outro, a probabilidade de mudança da posição do jogador em uma casa do tabuleiro para outra, é descrita pelo comportamento simultâneo de duas variaveis aleatórias, ou seja uma variavel aleatoria bidimensional, é a função de probabilidade conjunta entre elas. Considerando a variável aleatória X como o valor da face virada para cima dapós o lançamento do primeiro dado e a variável aleatória Y como o valor da face virada para cima dapós o lançamento do segundo dado, onde ambas possuem o mesmo espaço amostral Ω = {1,2,3,4,5,6}, temos que sua função de probabilidade, por definição, é p(xi; yj ) = P(X = xi; Y = yj ). A probabilidade de cada dado resultar cada possivel valor é de 1/6, uma vez que ambos são dados honestos.

Considere Y a variável aleatória que representa a quantidade de passos que o jogador dá até a sua primeira prisão. Para sermos capazes de obter uma aproximação do valor esperado de Y, podemos calcular a média a partir de varias simulações do algoritmo, ou seja reproduzindo varios jogos, somando a quantidade de passos que o jogador deu até sua primeira prisão em cada jogo e devidindo pelo número de jogos feitos. É importante pontuar que quanto maior o número de jogos, maior será a aproximação, pois teremos um número maior de simulações de um mesmo experiemento aleatorio e quanto mais jogadas, maior será a faixa de oportunidade que uma prisão aconteça, isto é deixo de limitar minha primeira prisão a valores mais baixos.

Tal raciocinio, se baseia na lei dos grandes números, que é um teorema fundamental da teoria da probabilidade que indica que, se repetirmos muitas vezes (tendendo ao infinito) o mesmo experimento, a frequência de um determinado evento acontecendo tende a ser uma constante. Ou seja, ao repetirmos o jogo com um mesmo número de jogadas varias e varias vezes estaremos nos aproximando de uma constante, que representa a frequencia com que o jogador é preso em um certo número de jogadas. Essa, por sua vez, será a probabilidade desse evento ocorrer.

```python
def jogodeterminado():
  j=10000
  soma=0
  casa=0
  jogadas=j
  jogo=0
  prisoes=0
  ant_prisao=0
  solto=1

  while(jogo!=jogadas and ant_prisao==0):
    jogo = jogo + 1
    if(solto==0):
      prisoes = prisoes + 1
    if(prisoes==1):
      ant_prisao=jogo
    while(solto==0):
      dado1 = random.randint(1,6)
      dado2 = random.randint(1,6)
      if(dado1==dado2):
        solto=1
      else:
        jogo = jogo + 1
        dado1 = random.randint(1,6)
        dado2 = random.randint(1,6)
        if(dado1==dado2):
          solto=1
        else:
          jogo = jogo + 1
          solto=1

    dado1 = random.randint(1,6)
    dado2 = random.randint(1,6)
    if(dado1==dado2):
      dado1 = random.randint(1,6)
      dado2 = random.randint(1,6)
      if(dado1==dado2):
        dado1 = random.randint(1,6)
        dado2 = random.randint(1,6)
        if(dado1==dado2):
          casa=20
          solto=0
        else:
          if((casa + dado1 + dado2) > 39):
            casa = casa + dado1 + dado2 - 40
          else:
            casa=casa + dado1 + dado2
      else:
        if(casa + dado1 + dado2 > 39):
          casa = casa + dado1 + dado2 - 40
        else:
          casa = casa + dado1 + dado2
    else:
      if(casa + dado1 + dado2 > 39):
          casa = casa + dado1 + dado2 - 40
      else:
          casa = casa + dado1 + dado2
  return ant_prisao
```

```python
def media(n):
  jogos=n
  soma=0
  while(jogos!=0):
    soma=soma+jogodeterminado()
    jogos=jogos-1
  media=soma/n
  print(media)
```
